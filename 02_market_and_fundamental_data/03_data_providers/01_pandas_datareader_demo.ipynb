## pandas-datareader for Market Data

### Yahoo Finance

```python
start = '2014'
end = datetime(2017, 5, 24)

yahoo= web.DataReader('FB', 'yahoo', start=start, end=end)
yahoo.head()
```

```python
mpf.plot(yahoo.drop('Adj Close', axis=1), type='candle')
plt.tight_layout()
```

### IEX

another data source.. 테스트 목적으로 몇번 했더니 이미 한도 넘었다 gg

```python
api_key = pd.read_csv('../../../data/api_key.csv')
IEX_API_KEY = api_key[api_key['api'] == 'IEX_API_KEY'].key.values[0]
```

```python
os.environ["IEX_API_KEY"] = IEX_API_KEY
IEX_API_KEY = getenv('IEX_API_KEY', None)
```

```python
start = datetime(2015, 2, 9)
# end = datetime(2017, 5, 24)

iex = web.DataReader('FB', 'iex', start, api_key=IEX_API_KEY)
iex.info()
```

```python
sns.set_style('whitegrid')
iex.close.plot(figsize=(14, 5))
sns.despine()
```

#### Book Data
