## Ticker object 쓰면 야후 파이낸스에 있는거 전체적으로 사용가능

```python
symbol = 'AAPL'
ticker = yf.Ticker(symbol) 

# Ticker 쓰면 한개 기업밖에 안된다 -> Tickers 쓰면 됨
# symbols = ['MSFT', 'AAPL', 'FB']
# tickers = yf.Tickers(symbols)

# 시계열 데이터만 받는거면 그냥 yf.download도 가능
# yf.download(tickers=symbols).stack()
```

```python
# 예를 들어 여기에 있는 Business Summary 이용해서 자연어처리 및 knowledge graph 작성 가능 (https://youtu.be/V_EEckDZIaY)
ticker.info
```

### Get market data

```python
yf.download(tickers=symbol, actions=True)
```

```python
# get_actions로 Dividends, Stock Splits 따로 부를 수 있다
data = ticker.history(period='3mo', interval='1d', start=None, end=None, actions=True, auto_adjust=True, back_adjust=False)
data
```
